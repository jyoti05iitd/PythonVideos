# Group by
- Group by and count
- Group by sum, mean
- Group by on multiple columns
- Group by and multiple aggregation

```python
sleep.head()
```

## Group by and count

```python
sleep.groupby(['BMI Category']).count()
```

```python
sleep.groupby(['BMI Category'])['Person ID'].count()
```

```python
sleep.groupby(['BMI Category'])['Person ID'].size()
```

```python
sleep.groupby(['BMI Category'])['Person ID'].agg(count = 'count')
```

## Group by and sum, mean

```python
sleep.groupby(['BMI Category']).sum()
```

```python
sleep.groupby(['BMI Category'])['Daily Steps'].sum()
```

```python
sleep.groupby(['BMI Category'])['Daily Steps'].agg(daily_steps_sum = 'sum')
```

```python
sleep.groupby(['BMI Category'])['Daily Steps'].agg(daily_steps_mean = 'mean')
```

```python
sleep.groupby(['BMI Category'])['Daily Steps'].agg(daily_steps_median = 'median')
```

## Group by on multiple columns

```python
sleep.groupby(['Gender', 'BMI Category'])['Sleep Duration'].agg(sleep_duration_mean = 'mean')
```

```python
sleep.groupby(['Occupation', 'BMI Category'])['Sleep Duration'].agg(sleep_duration_mean = 'mean')
```

```python
sleep.groupby(['Occupation', 'BMI Category'])['Sleep Duration'].agg(sleep_duration_mean = 'mean').sort_values(by = 'sleep_duration_mean', ascending = False)
```

```python
sleep.groupby(['Occupation', 'BMI Category'])['Sleep Duration']\
.agg(sleep_duration_mean = 'mean').round(2)\
.reset_index()\
.sort_values(by = ['Occupation', 'sleep_duration_mean', 'BMI Category'], ascending = [True, False, True])
```

## Group by and multiple aggregation

```python
sleep.groupby(['Occupation', 'BMI Category'])\
.agg(count=('Sleep Duration', 'count'), sleep_duration_mean=('Sleep Duration', 'mean')).round(2)\
.reset_index()\
.sort_values(by=['Occupation', 'sleep_duration_mean', 'BMI Category'], ascending=[True, False, True])
```
