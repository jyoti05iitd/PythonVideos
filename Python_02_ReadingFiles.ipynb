**Reading a csv file**

- Important arguements
    - filepath, 
    - sep, 
    - delimiter, 
    - header='infer', 
    - names, 
    - index_col=None, 
    - skipinitialspace=False, 
    - skiprows=None, 
    - skipfooter=0, 
    - nrows=None, 
    - na_values=None, 
    - parse_dates=None, 
    - infer_datetime_format, 
    - date_parser, 
    - date_format, 
    - dayfirst=False,
    - encoding=None

**Reading an excel file**
- pd.read_excel (https://pandas.pydata.org/docs/reference/api/pandas.read_excel.html)
- Important arguements
    - sheet_name=0,
    - header=0, 
    - names=None, 
    - index_col=None, 
    - usecols=None, 
    - dtype=None, 
    - skiprows=None, 
    - nrows=None, 
    - na_values=None

```python
sleep_excel = pd.read_excel("Data_From_Kaggle.xlsx", sheet_name=0, nrows = 5)
```

```python
sleep_excel.head(100)
```

```python
sleep_excel = pd.read_excel("Data_From_Kaggle.xlsx", sheet_name=1, nrows = 5)

sleep_excel.head(100)
```
