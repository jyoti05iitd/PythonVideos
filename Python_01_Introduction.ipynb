 **AND, OR, EXCLUSIVE OR**
* x & y - Truex if both x and y are True
* x | y - True if x or y are True
* x ^ y - True if x or y is True, but not both

```python
# x & y
print("(2>1)&(6==6) : AND condition is satisfied hence : ", (2>1)&(6==6))
             
# x | y
print("(2>1)|(6!=6) : OR condition is satisfied hence : ", (2>1)|(6!=6))

# x ^ y
print("(2>1)^(6!=6) : Exclusive - OR condition is satisfied hence : ", (2>1)^(6!=6))
```

**Tuple, List and Dict**
* Tuple - Cannot be modified once created
* List - List content can be modified
* Dict - Dictionary of key and value pair

```python
# tuple example

tup1 = 1,3,5
print(tup1)
```

```python
print(tup1[2])
```

```python
# nested tup

tup2 = (1,3,5),(2,4)
print(tup2)
```

```python
print(tup2[0][1])
```

```python
# unpacking
x,y = tup2
print(x)
print(y)
```

```python
# list example
list1 = [1,5,7,"odd"]
print(list1)
```

```python
# List functions

list1.append("even")
print("After append :", list1)

list1.insert(1, "inserted1")
print("After insertion at 1 : ", list1)
```

```python
# remove
list1.remove("odd")
print("After removal :", list1)
```

```python
# Checking values in a list
print("odd" in list1)

print("even" in list1)
```

```python
# concatenate
list1 + list([1,2])
```

```python
# sorting
list2 = list([1,6,8,2,5])
list2.sort()
print("Sorted list: ", list2)
```

```python
# zip two list

a = ["a", "b", "c", "d"]
x = ["x", "y", "z", "k"]

zip_ax = list(zip(a,x))
print(zip_ax)
```

```python
# unzip
a1, x1 = zip(*zip_ax)

print(a1)
print(x1)
```

**Dict**
* key-value pair
* check for a key
* delete using del or pop
* get keys using dict1.keys()
* get values using dict1.values()
* update using dict.update({'new_key': 'new_value'})

```python
# dict
dict1 = {"opener" : "Rahul", "keeper" : "Singh", "Bowler" : "Khan"}

print("Dictionary : ", dict1)

print("Check presence of opener in dict1 keys : ", "opener" in dict1)

print("Delete a key 'Bowler' with value : ", dict1.pop("Bowler"), " and After deletion : ", dict1)

print("All keys : ", dict1.keys())

print("All values : ", dict1.values())

print("Update dictionary with 'Bowler : Dev' : ", dict1.update({"bowler": "Dev"}), " And after update the dict is : ", dict1)
```
