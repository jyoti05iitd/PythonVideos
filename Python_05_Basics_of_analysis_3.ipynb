# Filter the dataframe for the following cases:
1. Gender = Male
2. Age = 27
3. Age > 50
4. Age between 30 & 35
5. Age > 40 and BMI Category is Obese
6. Daily Steps > 8000 & Sleep Disorder None
7. Daily Steps < 4000 or Sleep Disorder Insomnia
8. Occupation is either Doctor, Nurse or Engineer
9. Occupation is not Doctor
10. Occupation is not Doctor but age > 50

## Filter Gender =  Male

```python
sleep['Gender'].value_counts()
```

```python
sleep.loc[sleep['Gender'] == 'Male']
```

```python
sleep[sleep['Gender'] == 'Male']
```

## Filter Age = 27

```python
sleep[sleep['Age'] == 27]
```

## Filter Age > 50

```python
sleep[sleep['Age'] > 50]
```

## Filter Age between 30 & 35

```python
sleep[(sleep['Age'] >= 30) & (sleep['Age'] <= 35)]
```

## Filter Age > 40 and BMI Category is Obese

```python
sleep[(sleep['Age'] > 40) & (sleep['BMI Category'] == 'Obese')]
```

## Filter Daily Steps > 8000 & Sleep Disorder None

```python
sleep[(sleep['Daily Steps'] > 8000) & (sleep['Sleep Disorder'] == 'None')]
```

## Filter Daily Steps < 4000 or Sleep Disorder Insomnia

```python
sleep[(sleep['Daily Steps'] < 4000) & (sleep['Sleep Disorder'] == 'Insomnia')]
```

## Filter Occupation is either Doctor, Nurse or Engineer

```python
sleep['Occupation'].value_counts()
```

```python
sleep[(sleep['Occupation'] == 'Doctor')|(sleep['Occupation'] == 'Nurse')|(sleep['Occupation'] == 'Engineer')]
```

```python
Occ_selected = ['Doctor', 'Nurse', 'Engineer']

sleep[sleep['Occupation'].isin(Occ_selected)]
```

## Occupation is not Doctor

```python
sleep[sleep['Occupation'] != 'Doctor']
```

## Filter Occupation is not Doctor but age > 50

```python
sleep[(sleep['Occupation']!='Doctor')&(sleep['Age'] > 50)]
```
