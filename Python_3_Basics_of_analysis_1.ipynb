```python
import pandas as pd
```

```python
sleep = pd.read_csv("Data_From_Kaggle - Sleep_health_and_lifestyle_dataset.csv")
```

```python
sleep.shape
```

**Check the first few rows**

```python
sleep.head(10)
```

**Check the last few rows**

```python
sleep.tail(10)
```

**Get information about the data**

```python
sleep.info()
```

**Get summary of numeric columns**

```python
sleep.describe()
```

**Person ID should be string**

Changing the type to string

```python
sleep['Person ID'] = sleep['Person ID'].astype('str')
```

```python
sleep.describe()
```

**Selecting rows using iloc**

```python
# Starts with 0 but last value in array is not included
sleep.iloc[:2,]
```

```python
# Starts with 1 till 3
sleep.iloc[1:4,]
```

```python
# Last 5 rows are displayed
sleep.iloc[-5:,]
```

```python
sleep.iloc[[1,5,9],]
```

**Selecting columns**

```python
# Index starts from 0
# Columns 0 to 3 are selected i.e. 0,1,2,3
sleep.iloc[:5,:4]
```

```python
# Index starts from 1
# Columns 1 to 3 are selected i.e. 1,2,3
sleep.iloc[:5,1:4]
```

```python
# Last 3 columns are selected
sleep.iloc[:5,-3:]
```

```python
# Columns 0 to 1 and rows 0,1,2 are selected
sleep.iloc[[1,2],[0,3]]
```

**With loc and column names**

```python
# All the rows till index 5 are selected with loc
sleep.loc[:5,['Person ID','Occupation']]
```

```python
sleep.loc[:2,['Person ID', 'Occupation', 'Heart Rate']]
```

**Selecting with column names  - Select in SQL query**

```python
sleep[['Person ID', 'Occupation']]
```

**Selecting columns containing a term**

```python
sleep.columns
```

```python
sleep.columns.str.contains('Sleep')
```

```python
sleep.loc[:2,sleep.columns.str.contains('Sleep')]
```

```python
sleep.iloc[:2, [4,5,12]]
```

```python
sleep.loc[:2,sleep.columns.str.contains('ID|Sleep')]
```

**Selecting rows and columns**

```python
sleep.loc[:5,sleep.columns.str.contains('ID|Sleep')]
```

**Filtering rows and selecting columns**

```python
sleep.loc[sleep['Sleep Disorder'] == 'Insomnia',sleep.columns.str.contains('ID|Sleep')]
```
