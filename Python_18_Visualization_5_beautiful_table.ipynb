```python
df_summ = data.groupby('Occupation').agg(Mean_Sleep_Duration = ('Sleep Duration', 'mean'),
                                        Mean_Sleep_Quality = ('Quality of Sleep', 'mean'))\
.reset_index()\
.round(2)

df_summ
```

## Highlighting min/ max values

```python
import pandas as pd

# Define a function to apply conditional formatting
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]

def highlight_min(s):
    is_min = s == s.min()
    return ['background-color: lightblue' if v else '' for v in is_min]

# Apply the conditional formatting
df_summ_styled = df_summ.style.apply(highlight_max, subset=['Mean_Sleep_Duration', 'Mean_Sleep_Quality'])\
                              .apply(highlight_min, subset=['Mean_Sleep_Duration', 'Mean_Sleep_Quality'])

df_summ_styled
```

## Formatting of headers and a column

```python
import pandas as pd

# Apply the color gradient formatting
df_summ_styled = df_summ.style.background_gradient(subset=['Mean_Sleep_Duration', 'Mean_Sleep_Quality'], 
                                                   cmap='prism')

# Format the column names
df_summ_styled = df_summ_styled.set_table_styles(
    [{'selector': 'th', 'props': [('font-weight', 'bold'), ('background-color', '#000066'), 
                                 ('color', 'white'), ('font-size', '14px')]}]
)

# Format the Occupation column
df_summ_styled = df_summ_styled.set_properties(subset=['Occupation'], **{'font-weight': 'bold', 'color': 'white', 'background-color':'#000066', 'font-size': '14px'})

# Hide the index column
df_summ_styled = df_summ_styled.hide(axis='index')

df_summ_styled
```

## Formatting the headers, columns and values in the table

```python
import pandas as pd

# Apply the color gradient formatting with a very light color gradient
df_summ_styled = df_summ.style.background_gradient(subset=['Mean_Sleep_Duration', 'Mean_Sleep_Quality'], 
                                                   cmap='Greens', low=0.01, high=0.2)

# Format the column names
df_summ_styled = df_summ_styled.set_table_styles(
    [{'selector': 'th', 'props': [('font-weight', 'bold'), ('background-color', 'lightskyblue'), 
                                 ('color', 'black'), ('font-size', '14px')]}]
)

# Format the Occupation column
df_summ_styled = df_summ_styled.set_properties(subset=['Occupation'], **{'font-weight': 'bold', 'color': 'black', 'background-color':'lightskyblue', 'font-size': '14px'})

# Format the numeric columns to 2 decimal places
df_summ_styled = df_summ_styled.format({'Mean_Sleep_Duration': '{:.2f}', 'Mean_Sleep_Quality': '{:.2f}'})

# Hide the index column
df_summ_styled = df_summ_styled.hide(axis='index')

df_summ_styled
```

## Adding borders to the table with simple formatting

```python
import pandas as pd

# Remove the color gradient formatting and apply a solid background color
df_summ_styled = df_summ.style.set_properties(subset=['Mean_Sleep_Duration', 'Mean_Sleep_Quality'], 
                                              **{'background-color': 'lightyellow'})

# Format the column names
df_summ_styled = df_summ_styled.set_table_styles(
    [{'selector': 'th', 'props': [('font-weight', 'bold'), ('background-color', 'lightskyblue'), 
                                 ('color', 'black'), ('font-size', '14px')]}]
)

# Format the Occupation column
df_summ_styled = df_summ_styled.set_properties(subset=['Occupation'], **{'font-weight': 'bold', 'color': 'black', 'background-color':'lightskyblue', 'font-size': '14px'})

# Format the numeric columns to 2 decimal places
df_summ_styled = df_summ_styled.format({'Mean_Sleep_Duration': '{:.2f}', 'Mean_Sleep_Quality': '{:.2f}'})

# Hide the index column
df_summ_styled = df_summ_styled.hide(axis='index')

# Add table borders
df_summ_styled = df_summ_styled.set_table_styles(
    [{'selector': 'table', 'props': [('border', '2px solid black')]},
     {'selector': 'th', 'props': [('border', '4px solid red')]},
     {'selector': 'td', 'props': [('border', '2px solid black')]}],
    overwrite=False
)

# Define a function to apply conditional formatting
def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: lightgreen' if v else 'background-color: lightyellow' for v in is_max]


df_summ_styled = df_summ_styled.apply(highlight_max, subset=['Mean_Sleep_Duration', 'Mean_Sleep_Quality'])

df_summ_styled
```
